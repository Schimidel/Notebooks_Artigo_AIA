```python
gdf1 = gpd.read_file('50S_20S/50S_20S.kml', driver="KML")
gdf2 = gpd.read_file('25S_20N/25S_20N.kml', driver="KML")
```

```python
region = [-90, 50, -55, 30]
grid = pygmt.datasets.load_earth_relief(resolution="10m",  region=region)
```

```python
fig = pygmt.Figure()
fig.coast(region=region, projection="M0/0/015c", water="lightblue", land="lightgreen", borders=1)
fig.coast(shorelines=True)
fig.plot(gdf1, fill="red3", pen="0.5p,black",transparency=50)
fig.plot(gdf2, fill="blue3", pen="0.5p,black",transparency=50)
fig.basemap(frame=["ag", "+tAtlntico Sul"])
fig.text(text="Regio 1", x=-39, y=-35, font="13p,Helvetica-Bold,white")
fig.show()
```
